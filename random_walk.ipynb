```python
import numpy as np
import matplotlib.pyplot as plt
```

Simulate 10 steps:

```python
# Init
np.random.seed(99)
walks = []

# Simulate
for i in range(10):
    random_walk = [0]
    for _ in range(10):
        step = np.random.choice([-1, 1])
        random_walk.append(random_walk[-1] + step)
    walks.append(random_walk)

# Plot
plt.plot(walks)
plt.show()
```

Simulate 100 steps:

```python
# Init
np.random.seed(99)
walks = []

# Simulate
for i in range(20):
    random_walk = [0]
    for _ in range(100):
        step = np.random.choice([-1, 1])
        random_walk.append(random_walk[-1] + step)
    walks.append(random_walk)

# Convert
np_walks = np.array(walks)

# Plot np_walks
plt.plot(np_walks)
plt.show()

# Clear
plt.clf()

# Transpose 
np_walks_t = np.transpose(np_walks)

# Plot np_walks_t
plt.plot(np_walks_t)
plt.show()
```

Increase iterations:

```python
# Init
np.random.seed(99)
walks = []

# Simulate
for i in range(500):
    random_walk = [0]
    for _ in range(100):
        step = np.random.choice([-1, 1])
        random_walk.append(random_walk[-1] + step)
    walks.append(random_walk)

# Convert
np_walks = np.array(walks)

# Plot np_walks
plt.plot(np_walks)
plt.show()

# Clear
plt.clf()

# Transpose
np_walks_t = np.transpose(np_walks)

# Plot np_walks_t
plt.plot(np_walks_t)
plt.show()
```

Law of Large Numbers. Noticable normal distribution.

```python
final = np_walks_t[-1]
plt.hist(final)
plt.show()

# Calculate the odds
num = 0
bool = final >= num
greater = np.sum(bool)
n_walks = len(final)
print("The odds of the ending point being above " + str(num) + " is: ", greater/n_walks)
```
